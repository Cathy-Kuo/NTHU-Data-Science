```python
import requests
from bs4 import BeautifulSoup
import json
import re
from fake_useragent import UserAgent
```

```python
ua = UserAgent()
headers = {
    'User-Agent': 'Googlebot',
    'From': 'myth594@gmail.com'
}
title_a = []
rate_a = []
num_rate_a = []
num_sold_a = []
price_a = []
for i in range(1):
    print(i)
    front = 'https://search.taobao.tw/catalog/?_keyori=ss&from=input&page='
    page = str(i)
    back = '&q=%E8%83%8C%E5%BF%83%E5%A5%B3'
    url = front + page + back
    r = requests.get(url,headers=headers,allow_redirects=False)
    soup = BeautifulSoup(r.text, 'html.parser')
    pageConfig = soup.find('script', text=re.compile('window.pageData'))
    pagecon = pageConfig.string[16:len(pageConfig.string)]
    print(pagecon)
    data_json = json.loads(pagecon)
    for j in range(40):
        sour = 'http:'
        addr = sour + data_json['mods']['listItems'][j]['productUrl']
        r = requests.get(addr)
        soup = BeautifulSoup(r.text,"html.parser")
        title = soup.find('title').text
        buy = soup.find_all('script')[20]
        buy = buy.string[1328:len(buy.string)-331]
        buy_json = json.loads(buy)
        price = buy_json['data']['root']['fields']['price']['0']['salePrice']['value']
        rate = buy_json['data']['root']['fields']['product']['rating']['score']
        num_rate = buy_json['data']['root']['fields']['product']['rating']['total']
        if (len(buy_json['data']['root']['fields']['product']) == 9):
            l = len(buy_json['data']['root']['fields']['product']['middleText'])
            num_sold = buy_json['data']['root']['fields']['product']['middleText'][0:(l-4)]
        else:
            num_sold = 0
        title_a.append(title)
        rate_a.append(rate)
        num_rate_a.append(num_rate)
        num_sold_a.append(num_sold)
        price_a.append(price)
```
